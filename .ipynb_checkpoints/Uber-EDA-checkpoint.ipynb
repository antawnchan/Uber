# basic data analysis and cleaning

```python
city_metrics = pd.read_csv("./data_set_170705/city_metrics.csv")
driver_data = pd.read_csv("./data_set_170705/driver_data.csv")
driver_trips = pd.read_csv("./data_set_170705/driver_trips.csv")
rider_data = pd.read_csv("./data_set_170705/rider_data.csv")
rider_trips = pd.read_csv("./data_set_170705/rider_trips.csv")
```

```python
city_metrics["month"] = city_metrics["timestamp"].map(lambda t: t.month)
city_metrics["day"] = city_metrics["timestamp"].map(lambda t: t.day)
city_metrics["hour"] = city_metrics["timestamp"].map(lambda t: t.hour)
```

```python
city_metrics.groupby("hour").sum().to_csv("./city_metrics_hour.csv")
```

```python
city_metrics[city_metrics["hour"] == 0]
```

```python
city_metrics[city_metrics["hour"] == 0].plot(x="timestamp", y="trips")
```

```python
city_metrics["t_per_r"] = city_metrics.trips / city_metrics.requests
```

```python
ax = city_metrics.plot(x='timestamp', y='t_per_r', figsize=(7,4))
ax.set_xlabel("Date")
ax.set_ylabel("Trips per Request Ratio")
ax.set_title("Trips per Request Ratio over Date")
ax.legend("Trips Per Request")
```

```python

fig, ax = plt.subplots(1, 1)
city_metrics.groupby(["hour"]).mean().plot(y="t_per_r", ax=ax, figsize=(7,4), linewidth=4, color='b')
# ax.set_xlabel("Date")
ax.set_ylabel("Trips per Request Ratio")
ax.set_title("Trips per Request Ratio per Hour")
ax.legend("Trips Per Request")
```

```python

fig, ax = plt.subplots(1, 1)
city_metrics.groupby(["month","day"]).mean().plot(y="t_per_r", ax=ax, figsize=(12,7), linewidth=4, color='b')
# ax.set_xlabel("")
ax.set_ylabel("Time to Arrival (minutes)")
ax.set_title("Time of Arrival over Date")
ax.legend(["Estimated CC","Actual CC","Estimated ~CC","Actual ~CC"])
```

```python
ax = city_metrics.groupby(["month","day"]).sum()[["trips","requests"]].plot()
ax.set_ylabel("Trips")
ax.set_title("Trips per Day")
```
